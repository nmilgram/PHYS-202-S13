$V(x,y)=\frac{kq}{\sqrt{x^2 +(y-d)^2}} - \frac{kq}{\sqrt{x^2 +(y+d)^2}}$

```python
dx=0.06
dy=0.06
x=arange(-2.,2.,dx)
y=arange(-2.,2.,dy)
print x.ndim
```

```python
X,Y=meshgrid(x,y)
print X.ndim
print Y.shape
print X
print Y
```

```python
k=1
q=1
d=1

V=(k*q/(X**2+(Y-d)**2)**(1./2.))-(k*q/(X**2+(Y+d)**2)**(1./2.))
```

```python
print V
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig=figure()
ax=fig.gca(projection='3d')
p=ax.plot_surface(X,Y,V,rstride=1, cstride=1, linewidth=0, cmap=cm.copper)
```

```python
levels=array((6,2,1,.5,0.25,0,-0.25,-.05,-1,-2,-6))

contour(X,Y,V,levels,linewidths=3,cmap=cm.coolwarm)
show()
```

$G(x,y)=e^{(-x^2 -y^2)}$

```python
dx=0.06
dy=0.06
x=arange(-6.,6.,dx)
y=arange(-6.,6.,dy)
```

```python
X,Y=meshgrid(x,y)
G=e**(-(X**2)-(Y**2))
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig=figure()
ax=fig.gca(projection='3d')
p=ax.plot_surface(X,Y,G,rstride=1, cstride=1, linewidth=0, cmap=cm.jet)
```

```python
levels=array((0,.2,.4,.6,.8,1))

contour(X,Y,G,levels,linewidths=3,cmap=cm.copper)
show()
```

$V(x,y)=\frac{kq}{\sqrt{x^2 +(y-d)^2}} + \frac{kq}{\sqrt{x^2 +(y+d)^2}} - \frac{kq}{\sqrt{(x-d)^2 +y^2}} - \frac{kq}{\sqrt{(x+d)^2 +y^2}}$

```python
dx=0.06
dy=0.06
x=arange(-2.,2.,dx)
y=arange(-2.,2.,dy)
```

```python
X,Y=meshgrid(x,y)
```

```python
k= 8.987551787 #Nm^2 / C^2
q= 10**-9 #coulombs
d=1 #meters
V=(k*q/(X**2+(Y-d)**2)**(1./2.))+(k*q/(X**2+(Y+d)**2)**(1./2.)) - (k*q/(Y**2+(X+d)**2)**(1./2.)) - (k*q/(Y**2+(X-d)**2)**(1./2.))
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig=figure()
ax=fig.gca(projection='3d')
p=ax.plot_surface(X,Y,V,rstride=1, cstride=1, linewidth=0, cmap=cm.copper)
```

```python
levels=array(((1/4)*(10**-7),(1/3)*(10**-7), (1/2)*(10**-7),(10**-7),0,-(10**-7), -(1/2)*(10**-7), -(1/3)*(10**-7), -(1/4)*(10**-7)))
contour(X,Y,V,levels,linewidths=3,cmap=cm.coolwarm)
show()
```
