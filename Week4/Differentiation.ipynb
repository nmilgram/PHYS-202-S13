```python
x=arange(0., 10.1, 0.1)
y=x**5
```

```python
plot(x,y)
```

```python
dy =diff(y) #y(n+1) - y(n)
dx=diff(x)
dydx=dy/dx
print len(dydx)
```

```python
dydx1=zeros(y.shape,float)
dydx1[:-1] = diff(y)/diff(x)
dydx1[-1] = (y[-1] - y[-2]) / (x[-1] - x[-2])
```

$y=x^5, \frac{dy}{dx} = 5x^4$

```python
dydxExact = 5*x**4
plot(x, dydx1,x,dydxExact)
```
