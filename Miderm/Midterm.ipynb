```python
theta=arange(-3.14,3.14,.1)
```

```python
plot(theta, y(theta), color='red')
xlabel('$\theta(rad)$')
ylabel('intensity')
```

```python
def r(t):
    return (1+t)/(1-t)
def e(t):
    return exp(-t)
```

```python
t=arange(-5, 5, .01)
plot (t, -r(t), label='(1+t)/(1-t)')
plot (t, e(t), color='red', linestyle=':', label='exp(-t)')
pylab.ylim([-10,10])
pylab.xlim([-4.75,4.75])
xlabel('time(seconds)')
ylabel('f(t), g(t)')

legend()
show()
```

```python
def is_anagram(s1, s2):
    a=list(s1).sort()
    b=list(s2).sort()
    if a==b:
        return True
    else:
        return False
```
