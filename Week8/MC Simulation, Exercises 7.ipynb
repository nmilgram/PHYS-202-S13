c.)

```python
B=6
a=1.5

N=10000

z=numpy.random.random_sample(N);

Finv=lambda z: B*(-log(1-z))**(1/a)
    
xsamples=Finv(z)
```

d.)

```python
x=arange(0.,3.01, .01)

hist(xsamples, 50, normed=True, label='Sampled Distribution')

f=lambda x:a*(B**-a)*(x**(a-1))*(e**(-(x/B)**a))
    
F=lambda x:1 - (e**(-(x/B)**a))

plot(x, f(x), label='True Distribution', color='red',lw=3)
plot(x, F(x), label='CDF', color='g',lw=3)
xlim(0,15)
legend(loc=0)
show()
```

```python
def distribute1D(x,prob,N):
    
    f=prob
    
    F=cumsum(f(x))/sum(f(x))
    
    z=numpy.random.random_sample(N);

    Index= F.searchsorted(z)
    
    n,bins,patches = hist(x[Index],50, normed=True, label='Sampled Distribution')

    plot(x, f(x)*max(n), label='True Distribution', color='red',lw=3)
    
    xlabel('x')
    ylabel('Probability')
    
    legend(loc=0)
    show()
```

```python
distribute1D(arange(0.,5.01, .01),lambda x: 60*(x**3)*(1-x)**2 ,10000)
```
