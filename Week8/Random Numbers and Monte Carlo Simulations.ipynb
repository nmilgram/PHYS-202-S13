```python
x=arange(0.,pi+.01, .01)
f=lambda x: sin(x)/2.
F=lambda x:(cos(0)-cos(x))/2.
    
plot(x, f(x), label='True Distribution', lw=3)
plot(x, F(x), label='Cumulative Function', lw=3)
xlim(0,3.5)
ylim(0,1.1)
legend(loc=0)
show()
```

```python
N=10000

z=numpy.random.random_sample(N);

Finv=lambda z: arccos(cos(0)-2*z)
    
xsamples=Finv(z);

hist(xsamples, 50, normed=True, label='Sampled Distribution')

plot(x, f(x), label='True Distribution', color='red',lw=3)
xlim(0,pi)
ylim(0,.7)
legend(loc=0)
show()
```

$F(x)=-e^{-\lambda x}$

```python
x=arange(0.,5.01, .01)
f=lambda x: e**(-x)
F=lambda x:-e**(-x)
    
plot(x, f(x), label='True Distribution', lw=3)
plot(x, F(x), label='Cumulative Function', lw=3)
xlim(0,5.)
ylim(-1.,1.)
legend(loc=0)
show()
```

```python
N=10000

z=numpy.random.random_sample(N);

Finv=lambda z: -log(z)
    
xsamples=Finv(z);

hist(xsamples, 50, normed=True, label='Sampled Distribution')

plot(x, f(x), label='True Distribution', color='red',lw=3)

xlim(0,5.)
ylim(0,1.)
legend(loc=0)
show()
```

```python
x=arange(0.,5.01, .01)
f=lambda x: e**(-x)
F=cumsum(f(x))/sum(f(x))

plot(x, f(x), label='True Distribution', lw=3)
plot(x, F, label='Cumulative Sum', lw=3)
xlim(0,5.)
ylim(0.,1.)
legend(loc=0)
show()

```

```python
N=10000

z=numpy.random.random_sample(N);

Index= F.searchsorted(z[:])

hist(x[Index],50, normed=True, label='Sampled Distribution')

plot(x, f(x), label='True Distribution', color='red',lw=3)
xlim(0,5.)
ylim(0,1.)
legend(loc=0)
show()
```

```python
def distribute1D(x,prob,N):
    
    f=prob
    
    F=cumsum(f(x))/sum(f(x))
    
    z=numpy.random.random_sample(N);

    Index= F.searchsorted(z[:])
    
    hist(x[Index],50, normed=True, label='Sampled Distribution')

    plot(x, f(x), label='True Distribution', color='red',lw=3)
    xlim(0,5.)
    ylim(0,1.)
    legend(loc=0)
    show()
```

```python
distribute1D(arange(0.,5.01, .01),lambda x: e**(-x),10000)
```
